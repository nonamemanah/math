**Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.**

Имеется две точки x1 и y1. Для ортогональном преобразования этих дух точек, просто перенесем эти две точки на 5. Для доказательства того, что при ортогональном преобразовании сохраняется расстояние между точками, нам надо убедится, что расстояние между x1 и y1 такое же, как и растояние между x2 и y2.

```python
x1 = [1, 2]
y1 = [2, 1]
d1 = np.sqrt((y1[0] ** 2 - x1[0] ** 2) + (y1[1] ** 2 - x1[0] ** 2))
```

```python
x2 = 5 * x1
y2 = 5 * y1
d2 = np.sqrt((y2[0] ** 2 - x2[0] ** 2) + (y2[1] ** 2 - x2[0] ** 2))
```

```python
print(d1)
print(d2)
```

```python
d2 == d1
```

Так как d2 и d1 равны, то расстояние между точками сохраняется. Ч.т.д.
